## 4. Show Images From Drive

```python
data_path = "/content/drive/MyDrive/Data/Raw"

images = [files for files in os.listdir(data_path) if files.endswith(".jpg") or files.endswith(".jpeg")]
num = 6

for i in range(min(num, len(images))):
  image_path = os.path.join(data_path, images[i])

  img = cv2.imread(image_path)
  img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
  plt.imshow(img)
  plt.axis("off")
  plt.show()
```

## 5. Crop Image

```python
path = "/content/drive/MyDrive/Data/Raw/3.jpeg"

img = cv2.imread(path)

x, y, w, h = 100, 50, 300, 200

cropped = img[y:y+h, x:x+w]

plt.imshow(cv2.cvtColor(cropped, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
plt.figure(figsize=(8, 4))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(cropped, cv2.COLOR_BGR2RGB))
plt.title('Cropped Image')

plt.show()
```

## 7. Rotate Image

```python
rotated = cv2.rotate(img, cv2.ROTATE_90_COUNTERCLOCKWISE)
plt.imshow(cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB))
plt.show
```

## 8. Flip Image

```python
flipped_img = cv2.flip(img, -1)
plt.imshow(cv2.cvtColor(flipped_img, cv2.COLOR_BGR2RGB))
plt.show()
```
